# Task - 1 (5 points)
We want to use logistic regerssion to perdict if the patients will have heart problems or not. The column "Target" in our datasets includes data about heart diseases. If the patient had heart disease we have a 1 and if not a zero. 

Prepare your data set for predicting heart disease ("Target" column) out of 3 features:

* Age of the patient (Column **"Age"**)
* Gender of the patient (male or female - Column **"Sex"**)
* Cholestrol level of the patient (Column **"Chol"**) 


Split your data into 80% traning data and 20% test data. 

```python
# Adding additional libraries
from sklearn.model_selection import train_test_split

# Load dataset
heart_df = pd.read_csv("Heart.csv")

# Display the first few rows
heart_df.head()

# Selecting only the required features
X = heart_df[['Age', 'Sex', 'Chol']]  # Features
y = heart_df['Target']  # Target variable (0 = No heart disease, 1 = Heart disease)

# Splitting the dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Display the shapes of training and test sets
print(f"Training set: X_train = {X_train.shape}, y_train = {y_train.shape}")
print(f"Test set: X_test = {X_test.shape}, y_test = {y_test.shape}")
```

# Task 2 - (10 points)

Generate a logistic regression model using your training data. 

Print out the accuracy if your model. 

```python
# Additional librearies
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initializes the logistic regression model
log_reg = LogisticRegression()

# Trains the model using the training data
log_reg.fit(X_train, y_train)

# Predicts the target values for the test set
y_pred = log_reg.predict(X_test)

# Computes accuracy
accuracy = accuracy_score(y_test, y_pred)

# Prints the accuracy
print(f"Logistic Regression Model Accuracy: {accuracy:.2f}")
```

# Task 3 - (5 points)


Generate the classification report for your Logistic regresion model and interpret your results regarding precision, recall and f1-score. 


```python
from sklearn.metrics import classification_report

# Generate classification report
report = classification_report(y_test, y_pred)

# Print the report
print("Classification Report:\n", report)
```

The logistic regression model achieved **57% accuracy**, with a **precision of 55% for "No" (no heart disease) and 61% for "Yes" (heart disease)**. The recall for heart disease cases is **only 53%**, meaning **47% of actual heart disease cases were misclassified**, which is concerning for medical predictions.
